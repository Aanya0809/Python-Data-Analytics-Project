```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('C:\\Users\\aanya\\Downloads\\train.csv\\train.csv')
```

```python
del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.groupby('Age').size().plot(kind = 'bar', figsize = (12, 6), title = 'Purchase Distribution by Age')
```

```python
data = df.groupby('Age')['Product_ID'].nunique().reset_index()

data.columns = ['Age', 'Products']
```

```python
data.plot.bar(x = 'Age', figsize = (8,6))
```

```python
df['Purchase'] = pd.to_numeric(df['Purchase'], errors='coerce')
```

```python
purchase_sum_by_age = df.groupby('Age')['Purchase'].sum()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(purchase_sum_by_age.index, purchase_sum_by_age.values)
plt.xlabel('Age')
plt.ylabel('Total Purchase')
plt.title('Amount Spend by Age')
plt.show()
```

```python
purchase_mean_by_age = df.groupby('Age')['Purchase'].mean()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(purchase_mean_by_age.index, purchase_mean_by_age.values)
plt.xlabel('Age')
plt.ylabel('Total Purchase')
plt.title('Avg Amount Spend by Age')
plt.show()
```

```python

plt.pie(purchase_mean_by_age, labels=purchase_mean_by_age.index, autopct="%0.1f")

plt.title('Avg Amount Spend by Age')
plt.show()
```
