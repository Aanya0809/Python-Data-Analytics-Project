```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('C:\\Users\\aanya\\Downloads\\train.csv\\train.csv')
```

```python
del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

```python
sns.countplot(x = df['Stay_In_Current_City_Years'])
```

```python
df.groupby('Stay_In_Current_City_Years').size().plot(kind = 'pie', autopct = "%.1f")
```

```python
sns.countplot(x = 'Stay_In_Current_City_Years', hue = 'Gender', data = df)
```

```python
sns.countplot(x = 'Stay_In_Current_City_Years', hue = 'Marital_Status', data = df)
```

```python
sns.countplot(x = 'Stay_In_Current_City_Years', hue = 'City_Category', data = df)
```

```python
sns.countplot(x = 'City_Category', hue = 'Age',data = df)
```

```python
df['Purchase'] = pd.to_numeric(df['Purchase'], errors='coerce')
```

```python
purchase_by_stay_in_city=df.groupby('Stay_In_Current_City_Years')['Purchase'].sum()
```

```python
plt.title('Purchase amount by stay in the city')
plt.xlabel('Stay_In_Current_City_Years')
plt.ylabel('Purchase_Amount')
plt.bar(purchase_by_stay_in_city.index,purchase_by_stay_in_city.values)
```

```python
Mean_purchase_by_stay_in_city=df.groupby('Stay_In_Current_City_Years')['Purchase'].mean()
```

```python
plt.title('Mean Purchase amount by stay in the city')
plt.xlabel('Stay_In_Current_City_Years')
plt.ylabel('Purchase_Amount')
plt.bar(Mean_purchase_by_stay_in_city.index,Mean_purchase_by_stay_in_city.values)
```
