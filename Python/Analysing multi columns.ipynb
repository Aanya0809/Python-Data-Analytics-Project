```python
df['Purchase'] = pd.to_numeric(df['Purchase'], errors='coerce')
```

```python
purchase_sum_by_city = df.groupby('City_Category')['Purchase'].sum()
```

```python
plt.pie(purchase_sum_by_city, labels=purchase_sum_by_city.index, autopct="%0.1f")

plt.title('Amount Spend by City')
plt.show()
```

```python
purchase_mean_by_city = df.groupby('City_Category')['Purchase'].mean()
plt.pie(purchase_mean_by_city, labels=purchase_mean_by_city.index, autopct="%0.1f")
plt.title('Avg Amount Spend by City')
plt.show()
```
