```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('C:\\Users\\aanya\\Downloads\\train.csv\\train.csv')
```

```python
del df['Product_Category_2']
del df['Product_Category_3']
```

```python
gender_counts = df['Gender'].value_counts()
plt.title('Gender Ratio')
plt.pie(gender_counts, labels=gender_counts.index,radius=1,autopct="%.1f")
```

```python
plt.figure(figsize=(6,6))
plt.xlabel('Gender')
plt.ylabel('Number of buyers')

plt.bar(gender_counts.index,gender_counts.values,width=0.5)
```
